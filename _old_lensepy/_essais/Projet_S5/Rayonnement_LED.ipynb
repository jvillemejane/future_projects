Affichage 3D / Source LED sur un plan

```python
import numpy as np
def I_led_gauss(I0, half_angle, theta):
    """
    I0 : intensité sur l'axe de la source
    half_angle : demi-angle d'émission de la source en degre
    theta : angle pour lequel on souhaite l'intensité
    """
    return (I0 * np.exp(-(4*np.log(2)) * (theta/(2*half_angle))**2))
```

```python
I0 = 1 # Intensité sur l'axe
angle_direction = 30
```

Diagramme de rayonnement de la source

```python
theta = np.linspace(0, 180, 181)
i_led = I_led_gauss(I0, angle_direction, theta)
```

```python
from matplotlib import pyplot as plt
plt.figure()
plt.plot(theta, i_led)
```

Affichage polaire

```python
fig = plt.figure()
ax = fig.add_subplot(projection='polar')
ax.plot(theta/180*np.pi, i_led)
ax.set_thetamin(0)
ax.set_thetamax(90)
ax.set_theta_direction(-1)
ax.set_theta_offset(np.pi / 2.0)

plt.show()
```
